```python
all_data = datasets.list(limit=3)
```

```python
my_dataset = all_data[0]
images = my_dataset.images()
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(15, 15))
columns = 2
rows = round((len(images) + 1)/columns)
i = 1
for img in images:
    fig.add_subplot(rows, columns, i)
    plt.imshow(img.img)
    i += 1
plt.show()
```
